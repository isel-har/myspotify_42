Collaborative filtering approach

User-based CF top 10 recommendations and model evaluation

```python
from tools.cf_rec import CFRecommender

rec = CFRecommender()

train_data, test_data = rec.load_split()

user_item_matrix = rec.load_sparse()

rec.load_mapping()

model = rec.load_model()

print("CF requirements loaded")
```

```python
rec.evaluate_model(model, test_data, user_item_matrix, n_users=100)
```

```python
user_id = 'b7815dbb206eb2831ce0fe040d0aa537e2e800f7'
# user_id = '02a4255f067037ab82375a12d941a3df6ba93248'

song_df = rec.load_data(song_df=True)

rec_df, precision = rec.similar_to_you_listen(
    user_id=user_id,
    song_df=song_df,
    test_data=test_data,
    model=model,
    user_item_matrix=user_item_matrix
)

print(f"precision at 10 for this recommendation: {(precision * 100):.2f} %")
rec_df
```

Item-based CF model evaluation

```python
avg_precision = rec.average_precision_at_k_given_track(
    song_id='SOAUWYT12A81C206F1',
    train_data=train_data,
    test_data=test_data,
    song_df=song_df,
    model=model,
    K=10
)
```

top 10 recommendations with likelihood

```python
rec_df = rec.get_similar_songs(
    song_id='SOWYSKH12AF72A303A',
    train_data=train_data,
    song_df=song_df,
    model=model
)

user_actual = set(test_data[test_data.user_id == user_id]['song_id'])
precision = rec.calulate_precision(rec_df['song_id'], user_actual, 10)

print(f"precision of this 10 recommendations: {(precision*100):.2f}")
rec_df[['artist', 'title', 'likelihood']]
```
